```python
# 딕셔너리는 어떻게 만들까?
dic = {"name" : "json", "phone" : "010-1234-1234", "birth": "00-01-01"}
```

```python
# 딕셔너리 쌍 추가, 삭제하기
dic["hobby"] = "running"
```

```python
dic[3] = [1,2,3]
```

```python
# 딕셔너리 요소 삭제하기
del dic[3]
dic
```

```python
# 딕셔너리 요소 변경하기
dic["name"] = "python"
dic
```

```python
# 딕셔너리에서 Key를 사용해서 Value 얻기
dic["name"]
```

```python
print(dic["name"])
print(dic["phone"])
print(dic["birth"])
```

```python
# 딕셔너리 관련 함수
## Key 리스트 만들기 - keys

dic = {"name" : "json", "phone" : "010-1234-1234", "birth": "00-01-01"}

dic.keys()
list(dic.keys())
```

```python
# value 리스트 만들기 - values
li = dic.values()
list(li)
```

```python
# Key, Value 쌍 얻기 - items
li = dic.items
list(li)
```

```python
# key : value 쌍 모두 지우기 - clear
dic.clear()
dic
```

```python
# Key로 Value 얻기 - get

dic = {"name" : "json", "phone" : "010-1234-1234", "birth": "00-01-01"}

dic["name"]
print(dic.get("name"))
print(dic.get("phone"))
print(dic.get("birth"))
```

```python
# 해당 key가 딕셔너리 안에 있는지 조사하기 -in

dic = {"name" : "json", "phone" : "010-1234-1234", "birth": "00-01-01"}

"name" in dic
```

```python
# Key로 Value 얻기 -pop
dic = {"name" : "json", "phone" : "010-1234-1234", "birth": "00-01-01"}

phone = dic.pop("phone")
print(phone)
```
