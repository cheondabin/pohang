```python
money = True
if money: print("택시를 타고 가라.")
else: print("걸어가라.")
```

```python
# if문의 기본구조
# if 조건문: 
#     수행할 문장1
#     수행할 문장2
#     ...
# else:
#     수행할 문장1
#     수행할 문장2
#     ...
```

```python
# 학점이 90 이상이면 A, 80이상 90미만이면 B, 70이상 80미만이면 C, 60이상 70미만이면 D, 60미만이면 촐력하는 프로그램을 만드시오.
score = int(input("학점을 입력하세요:"))

if score >= 90:
   print(f"입력값은 {score} 당신의 학점은: A")
elif score >=80:
     print(f"입력값은 {score} 당신의 학점은: B")
elif score >=70:
     print(f"입력값은 {score} 당신의 학점은: C")
elif score >=60:
     print(f"입력값은 {score} 당신의 학점은: D")
else:
     print(f"입력값은 {score} 당신의 학점은: F")


   
```
